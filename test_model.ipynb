```python
import cv2
import matplotlib.pyplot as plt
from keras.models import load_model


class_names = ['MildDemented', 'ModerateDemented', 'NonDemented', 'VeryMildDemented']

model = load_model('/content/drive/MyDrive/model.hdf5')


image = cv2.imread('/content/drive/MyDrive/data/OriginalDataset/ModerateDemented/27 (2).jpg')
resized_image = cv2.resize(image, (176, 176))


x_test = resized_image.reshape(1, 176, 176, 3)
predictions = model.predict(x_test)


predicted_class_index = predictions.argmax(axis=-1)[0]

predicted_class_name = class_names[predicted_class_index]

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Input Image')
plt.show()

print('Predicted Class:', predicted_class_name)

```
